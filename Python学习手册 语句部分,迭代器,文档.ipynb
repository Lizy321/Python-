```python
def min1(*args):
    res = args[0]
    for arg in args[1:]:
        if arg<res:
            res = arg
    return res
def min2(first, *rest):
    print(first)
    print(rest)
    for arg in rest:
        print(arg)
        if arg<first:
            first = arg
    return first
def min3(*args):
    print(args)
    tmp = list(args)
    print(tmp)
    tmp.sort()
    return tmp[0]

print(min1(3,4,1,2),'\n')
print(min2('bb','aa'),'\n')
print(min3([2,2],[1,1],[3,3]))
```

```python
def minmax(test,*args):
    res = args[0]
    for arg in args[1:]:
        if test(arg,res):
            res = arg
    return res

def lessthan(x,y): return x<y
def grtrthan(x,y): return x>y

print(minmax(lessthan,4,2,1,5,6,3))
print(minmax(grtrthan,4,2,1,5,6,3))
```

```python
# set
def intersect(*args):
    res = []
    for x in args[0]:
        for other in args[1:]:
            if x not in other:
                break
        else: # break out of loop
            res.append(x)
    return res

def union(*args):
    res = []
    for seq in args:
        for x in seq:
            if not x in res:
                res.append(x)
    return res

s1,s2,s3 = 'spam','scam','slam'
print(intersect(s1,s2,s3))
print(union(s1,s2,s3))
```

### 递归函数

```python
def mysum(L):
    print(L)
    if not L:
        return 0
    else:
        return L[0] + mysum(L[1:])

mysum([1,2,3,4])
```

```python
def mysum1(L):
    return 0 if not L else L[0] + mysum1(L[1:])

def mysum2(L):
    return L[0] if len(L)==1 else L[0] + mysum2(L[1:])

def mysum3(L):
    first, *rest = L
    return first if not rest else first + mysum3(rest)
```

```python
#递归 求树的和
def sumtree(L):
    tot = 0
    for x in L:
        if not isinstance(x,list):
            tot += x
        else:
            tot += sumtree(x)
    return tot
    
L = [1,[2,[3,4],5],6,[7,8]]
print(sumtree(L))
```

```python
# indirect function
def echo(msg):
    print(msg)
    
schedule = [(echo,"a"),(echo,"b")]
for (func,arg) in schedule:
    func(arg)

# 对象模式，灵活性
def make(label):
    def echo(msg):
        print(label+':'+msg)
    return echo

F = make('spam')
F('Ham!')
F('Egg!')
```

### 生成器 yield / return

```python
# __next__()
# 内存使用和性能方面都更好

#生成器函数
def gen(n):
    for i in range(n):
        yield i**2

for i in gen(3):
    print(i, end=':')
    
print('')
x = gen(4)
print(next(x))
print(next(x))
print(next(x))
```

```python
# send(); next()
def gen1():
    for i in range(10):
        x = yield i**2
        print(x)
        
G = gen1()
next(G)

G.send(70)
G.send(60)
next(G)
```
